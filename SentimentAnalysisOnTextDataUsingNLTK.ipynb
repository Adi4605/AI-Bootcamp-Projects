```python
import nltk
from nltk.corpus import movie_reviews
from nltk.classify import NaiveBayesClassifier
from nltk.classify.util import accuracy as nltk_accuracy
from nltk.corpus import stopwords
import random
```

```python
nltk.download('movie_reviews')
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')
```

```python
def extract_ft(words):
  return{word:True for word in words}
```

```python
# Load movie_reviews dataset from NLTK
docs = [(list(movie_reviews.words(fileid)),category)
        for category in movie_reviews.categories()
        for fileid in movie_reviews.fileids(category)]
```

```python
# Shuffle dataset to ensure random distribution
random.shuffle(docs)
```

```python
ftsets = [(extract_ft(d),c) for (d,c) in docs]
train_set, test_set = ftsets[:1600], ftsets[1600:]
```

```python
classifier = NaiveBayesClassifier.train(train_set)
```

```python
accuracy = nltk_accuracy(classifier, test_set)
print(f'Accuracy: {accuracy*100:.2f}%')
```

```python
classifier.show_most_informative_features(10)
```

```python
def analyse_sentiment(text):
  # Tokenize and remove stopwords
  words = nltk.word_tokenize(text)
  words = [word for word in words if word.lower() not in stopwords.words('english')]
  # Extract features
  ft = extract_ft(words)
  # Predict sentiment
  sentiment = classifier.classify(ft)
  return sentiment
```

```python
test_sentences = [
    "This movie is absolutely fantastic! The acting, the story, everything was amazing!",
    "I hated this movie. It was a waste of time and money",
    "The plot was a bit dull, but the performances were great.",
    "I have mixed feelings about this film. It was okay, not great but not terrible ether."
]
```

```python
for s in test_sentences:
  print(f"Sentence: {s}")
  print(f"Sentiment: {analyse_sentiment(s)}")
  print()
```
